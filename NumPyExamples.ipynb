<h2>~Linear Algebra (np.linalg)</h2>

```python
import numpy as np

A = np.array([[2, 3],                     #Finds the determinant of a square matrix.
              [1, 4]])

print(np.linalg.det(A))
```

```python
import numpy as np

A = np.array([[1, 2],                                  #Finds the inverse of a square matrix.
              [3, 4]])

print(np.linalg.inv(A))
```

```python
import numpy as np

A = np.array([[2, 1],                                    #Solves the matrix equation Ax = b.
              [1, 3]])

b = np.array([8, 13])

x = np.linalg.solve(A, b)

print(x)
```

```python
import numpy as np

A = np.array([[2, 0],
              [0, 3]])

values, vectors = np.linalg.eig(A)                            #import numpy as np

A = np.array([[2, 0],
              [0, 3]])

values, vectors = np.linalg.eig(A)

print("Eigenvalues:")
print(values)

print("Eigenvectors:")
print(vectors)

```

```python
import numpy as np

A = np.array([[2, 3],                     #Finds the determinant of a square matrix.
              [1, 4]])

print(np.linalg.det(A))
```

```python
import numpy as np

A = np.array([[1, 2],                                  #Finds the inverse of a square matrix.
              [3, 4]])

print(np.linalg.inv(A))
```

```python
import numpy as np

A = np.array([[2, 1],                                    #Solves the matrix equation Ax = b.
              [1, 3]])

b = np.array([8, 13])

x = np.linalg.solve(A, b)

print(x)
```

```python
import numpy as np

A = np.array([[2, 0],
              [0, 3]])

values, vectors = np.linalg.eig(A)                            #import numpy as np

A = np.array([[2, 0],
              [0, 3]])

values, vectors = np.linalg.eig(A)

print("Eigenvalues:")
print(values)

print("Eigenvectors:")
print(vectors)

```

```python
import numpy as np

A = np.array([[1, 2],
              [2, 4]])

print(np.linalg.matrix_rank(A))                               #Returns the rank of a matrix.
```

<h2>Handling Missing Data</h2>

```python
import numpy as np

a = np.array([10, 20, np.nan, 40])                                 #Represents a missing value.

print(a)
```

```python
import numpy as np

a = np.array([10, np.nan, 30])                             #Checks whether each element is NaN.

print(np.isnan(a))
```

```python
import numpy as np

a = np.array([10, 20, np.nan, 40])                          #Calculates the mean while ignoring NaN values.

print(np.nanmean(a))
```

```python
import numpy as np

a = np.array([10, 20, np.nan, 40])                     #Calculates the sum while ignoring NaN values.

print(np.nansum(a))
```
