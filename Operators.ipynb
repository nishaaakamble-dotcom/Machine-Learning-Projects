```python
# Integer
num = 10
print(num)
print(type(num))
```

```python
# Float
price = 99.99
print(price)
print(type(price))
```

```python
# String
name = "Nisha"
print(name)
print(type(name))
```

```python
# Boolean
is_student = True
print(is_student)
print(type(is_student))
```

```python
# List
numbers = [10, 20, 30]
print(numbers)
print(type(numbers))
```

```python
# Tuple
colors = ("Red", "Green", "Blue")
print(colors)
print(type(colors))
```

```python

# Set
fruits = {"Apple", "Banana", "Mango"}
print(fruits)
print(type(fruits))
```

```python
# Dictionary
student = {"name": "Nisha", "age": 20}
print(student)
print(type(student))
```

```python
##list indexig
fruits = ["Apple", "Banana", "Mango", "Orange"]

print(fruits[0])   
print(fruits[1])   
print(fruits[2])   
print(fruits[3])   
```

```python
##nagative indexing
fruits = ["Apple", "Banana", "Mango", "Orange"]

print(fruits[-1])   # Orange
print(fruits[-2])   # Mango
print(fruits[-3])   # Banana
print(fruits[-4])   # Apple
```

```python
##list slicing
numbers = [10, 20, 30, 40, 50, 60]
print(numbers[0:3])
print(numbers[2:5])
print(numbers[2:])
print(numbers[:4])
print(numbers[::-1])
```

```python
##Arithmetic operators
a = 10
b = 3

print( a + b)
print( a - b)
print( a * b)
print( a / b)
print( a // b)
print( a % b)
print( a ** b)
```

```python
##compareason operators
a = 10
b = 20

print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
```

```python
##logical operator
x = True
y = False

print(x and y)
print(x or y)
print(not x)
```
