<h3>map , apply , DataFrame.map and Lambada Function</h3>

```python
import pandas as pd
import numpy as np

df=pd.DataFrame({

"Name":["Nisha","Sandyy","Khushii","Raj","Priya"],

"Age":[21,22,20,23,21],

"Gender":["Male","Female","Female","Male","Female"],

"Marks":[80,90,75,88,95],

"City":["Pune","Mumbai","Delhi","Pune","Nashik"]

})
print(df)
```

```python
# map using lambda

df["Age_Months"]=df["Age"].map(lambda x:x*12)
df
```

```python
# map using dictionary

gender={

"Male":1,

"Female":0

}

df["Gender_Code"]=df["Gender"].map(gender)
print(df)
```

```python
# apply on Series

df["Marks_Square"]=df["Marks"].apply(lambda x:x**2)
df
```

```python
# apply row-wise

def result(row):

    if row["Marks"]>=85:

        return "Excellent"

    else:

        return "Average"

df["Result"]=df.apply(result,axis=1)
df
```

<h3>~Missing Values</h3>

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Name":["Nisha","Sandyy","Khushii","Raj","Ravi"],
    "Age":[21,np.nan,20,23,np.nan],              
    "Marks":[80,90,np.nan,88,95],
    "City":["Pune","Mumbai",np.nan,"Pune","Nashik"]
})

print(df)
```

```python
print("Missing Values")
print(df.isnull().sum())    #Check missing values
```

```python
print("Missing Values")     #Count missing values
print(df.isnull().sum())
```

```python
print("After dropna()")     #Remove rows/columns with missing values
print(df.dropna())        
```

```python
print("Fill Age with Mean")
df["Age"] = df["Age"].fillna(df["Age"].mean())    #Fill numeric values with average
print(df)
```

```python
print("Fill Marks with Median")
df["Marks"] = df["Marks"].fillna(df["Marks"].median())    #Fill with median
print(df)
```

```python
print("Fill City with Mode")
df["City"] = df["City"].fillna(df["City"].mode()[0])       #Fill text values with the most common value
print(df)
```

```python
print("Interpolation Example")
s = pd.Series([10, np.nan, np.nan, 40, 50])    #Estimate missing values between known values
print(s.interpolate())
```

<h3>Complete Example</h3>

```python
import pandas as pd

df = pd.DataFrame({
    "Name":["Nisha","Sandyy","Khushii","Raj","Ravi","Rahulj"],
    "Age":[21,22,20,23,22,23],
    "Marks":[80,90,75,88,90,88],
    "City":["Pune","Mumbai","Delhi","Pune","Mumbai","Pune"]
})

print("Original Data")
print(df)

```

```python
# Check duplicates
print("Duplicate Rows")
print(df.duplicated())

```

```python
# Count duplicates
print("Total Duplicates")
print(df.duplicated().sum())

```

```python
# View duplicate rows
print("Duplicate Records")
print(df[df.duplicated()])

```

```python
# Duplicate based on Age
print("Duplicate based on Age")
print(df[df.duplicated(subset=["Age"])])

```

```python
# Remove duplicates
print("After Removing Duplicates")
print(df.drop_duplicates())

```

```python
# Keep last occurrence
print("Keep Last Duplicate")
print(df.drop_duplicates(keep="last"))

```

```python
# Remove every duplicate
print("Remove All Duplicates")
print(df.drop_duplicates(keep=False))
```
