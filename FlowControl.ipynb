1.	Check whether a number is even or odd.

```python
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")
```

2.	Check whether a number is positive, negative, or zero. 

```python
num = int(input("Enter a number: "))

if num > 0:
    print("Positive Number")
elif num < 0:
    print("Negative Number")
else:
    print("Zero")
```

3.	Find the largest of two numbers. 

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

if num1 > num2:
    print(num1, "is larger")
else:
    print(num2, "is larger")
```

4.	Find the largest of three numbers. 

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))

if num1 >= num2 and num1 >= num3:
    print(num1, "is largest")
elif num2 >= num1 and num2 >= num3:
    print(num2, "is largest")
else:
    print(num3, "is largest")
```

7. Find the Factorial of a Number

```python
num = int(input("Enter a number: "))
fact = 1

for i in range(1, num + 1):
    fact = fact * i

print("Factorial =", fact)
```

10. Find the Sum of Cubes of First N Natural Numbers

```python
n = int(input("Enter N: "))
sum = 0

for i in range(n + 1):
    sum = sum + i ** 3

print("Sum of Cubes =", sum)
```

11. Reverse a Number

```python
num = int(input("Enter a number: "))
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

print("Reverse =", rev)
```

12.	Check whether a number is palindrome.
   

```python
num = input("Enter a number: ")

if num == num[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

13.Given Number is Armstrong or Not

```python
num = int(input("Enter a number: "))
s = 0
temp = num
n = len(str(num))

while temp > 0:
    digit = temp % 10
    s = s + digit ** n
    temp = temp // 10

if s == num:
    print("Armstrong")
else:
    print("Not Armstrong")
```

14.Given number is Perfect or Not

```python
num = int(input("Enter a number: "))
s = 0

for i in range(1, num):
    if num % i == 0:
        s = s + i

if s == num:
    print("Perfect Number")
else:
    print("Not Perfect Number")
```

15.Sum of Digits of a Number

```python
num = int(input("Enter a number: "))
s = 0

while num > 0:
    s = s + num % 10
    num = num // 10

print("Sum of digits =", s)
```

17.Find the Product of Digits of a Number

```python
num = int(input("Enter a number: "))
p = 1

while num > 0:
    digit = num % 10
    p = p * digit
    num = num // 10

print("Product =", p)
```

18.Find the Largest Digit in a Number

```python
num = int(input("Enter a number: "))
largest = 0

while num > 0:
    digit = num % 10
    if digit > largest:
        largest = digit
    num = num // 10

print("Largest digit =", largest)
```

19.Find the Smallest Digit in a Number

```python
num = int(input("Enter a number: "))
smallest = 9

while num > 0:
    digit = num % 10
    if digit < smallest:
        smallest = digit
    num = num // 10

print("Smallest digit =", smallest)
```

20.Count the Number of Digits in a Number

```python
num = int(input("Enter a number: "))
count = 0

while num > 0:
    count = count + 1
    num = num // 10

print("Number of digits =", count)
```

21. Find the First and Last Digit of a Number

```python
num = int(input("Enter a number: "))
last = num % 10

while num >= 10:
    num = num // 10

first = num

print("First digit =", first)
print("Last digit =", last)
```

23.Find LCM of Two Numbers

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

lcm = max(a, b)

while True:
    if lcm % a == 0 and lcm % b == 0:
        print("LCM =", lcm)
        break
    lcm = lcm + 1
```
