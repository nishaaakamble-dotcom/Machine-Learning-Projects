```python
a = frozenset([1, 2, 3])
b = frozenset([3, 4, 5])

print(a.union(b))
```

```python
a = frozenset([1, 2, 3])
b = frozenset([2, 3, 4])

print(a.intersection(b))
```

```python
a = frozenset([1, 2, 3])
b = frozenset([2, 3, 4])

print(a.difference(b))
```
